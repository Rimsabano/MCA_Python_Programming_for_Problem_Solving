```python
import math

number = int(input("Enter a number: "))
power_value = int(input("Enter the power value: "))
```

```python
square_root = math.sqrt(number)
factorial_value = math.factorial(number)
power_result = math.pow(number, power_value)

print("Square root of the number:", square_root)
print("Factorial of the number:", factorial_value)
print("Power of the number:", power_result)
```
