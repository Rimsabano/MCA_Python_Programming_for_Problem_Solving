```python
for i in range(1, 6):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
rows = 5
for i in range(rows):
    print("  " * (rows - i - 1), end="")
    print("* " * (2*i + 1))
```

```python
rows = 3
for i in range(rows):
    print("  " * (rows - i - 1), end="")
    print("* " * (2*i + 1))
for i in range(rows-2, -1, -1):
    print("  " * (rows - i - 1), end="")
    print("* " * (2*i + 1))
```
