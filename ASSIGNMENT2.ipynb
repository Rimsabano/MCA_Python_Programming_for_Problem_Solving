```python
def dice_probability(number):
    if 1 <= number <= 6:
        return 1/6
    return 0
def coin_probability(side):
    side = side.lower()
    if side in ["heads", "tails"]:
        return 1/2
    return 0
```

```python
print("1. Probability of Dice Roll")
print("2. Probability of Coin Flip")
choice = int(input("Enter your choice (1/2): "))
```

```python
if choice == 1:
    num = int(input("Enter number on dice (1â€“6): "))
    p = dice_probability(num)
    if p > 0:
        print(f"Probability of rolling {num} = {p} ({p*100}%)")
    else:
        print("Invalid dice number!")
elif choice == 2:
    side = input("Enter side (Heads/Tails): ")
    p = coin_probability(side)
    if p > 0:
        print(f"Probability of getting {side.capitalize()} = {p} ({p*100}%)")
    else:
        print("Invalid coin side!")
else:
    print("Invalid choice!")
```
