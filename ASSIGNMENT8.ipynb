```python
import os
```

```python
directory_files = ['report.pdf', 'data.csv', 'photo.jpg', 'summary.pdf', 'script.py']
```

```python
file_dict = {}
for file in directory_files:
    name, ext = os.path.splitext(file)
    ext = ext[1:]  
    if ext in file_dict:
        file_dict[ext].append(file)
    else:
        file_dict[ext] = [file]
file_dict
```
