```python
num = int(input("Enter a number: "))
print("You entered:", num)
```

```python
binary_value = bin(num)[2:]
octal_value = oct(num)[2:]
hex_value = hex(num)[2:]
print("Conversions completed.")
```

```python
print("Binary:", binary_value)
print("Octal:", octal_value)
print("Hexadecimal:", hex_value)
```
