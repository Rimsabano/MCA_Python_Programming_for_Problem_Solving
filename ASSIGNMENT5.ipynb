```python
employees = [
    {"name": 'John', "salary": 45000},
    {"name": 'Mary', "salary": 52000},
    {"name": 'Alex', "salary": 48000}
]
employees
```

```python
sorted_employees = sorted(employees, key=lambda x: x['salary'])
sorted_employees
```

```python
print("Sorted by Salary:", sorted_employees)
```
