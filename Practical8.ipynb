```python
data = input("Enter student names and marks (format: Name:Marks, Name:Marks): ")
students = []
for pair in data.split(","):
    name, marks = pair.strip().split(":")
    students.append((name, int(marks)))
print("You entered:", students)
```

```python
topper = max(students, key=lambda x: x[1])
print("Topper identified.")
```

```python
print(f"Topper: {topper[0]} ({topper[1]})")
```
