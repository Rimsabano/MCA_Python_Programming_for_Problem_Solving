```python
text = input("Enter sentence: ")
print("You entered:", text)
```

```python
vowels = "aeiouAEIOU"
only_vowels = [ch for ch in text if ch in vowels]
only_vowels.reverse()
print("Reversed vowels list:", only_vowels)
```

```python
result = ""
vowel_index = 0
for ch in text:
    if ch in vowels:
        result += only_vowels[vowel_index]
        vowel_index += 1
    else:
        result += ch
print("Result:", result)
```
