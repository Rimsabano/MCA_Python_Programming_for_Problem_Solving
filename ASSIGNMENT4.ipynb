```python
numbers = [12, 45, 33, 21, 88]
item_to_search = 33
numbers, item_to_search
```

```python
position = -1
for index, value in enumerate(numbers, start=1):  # start=1 for 1-based position
    if value == item_to_search:
        position = index
        break
position
```

```python
if position != -1:
    print(f"Item found at position {position}")
else:
    print("Item not found.")
```
