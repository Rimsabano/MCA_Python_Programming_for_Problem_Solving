```python
import csv
with open("employees.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["John", "IT", 50000])
    writer.writerow(["Mary", "IT", 55000])
    writer.writerow(["Alice", "HR", 48000])
    writer.writerow(["Bob", "HR", 52000])

print("CSV file 'employees.csv' created successfully!")
```

```python
import csv

dept_data = {}
with open("employees.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        name, dept, salary = row
        salary = float(salary)
        if dept in dept_data:
            dept_data[dept].append(salary)
        else:
            dept_data[dept] = [salary]

print("Average Salary by Department:")
for dept, salaries in dept_data.items():
    avg_salary = sum(salaries) / len(salaries)
    print(f"{dept}: {avg_salary}")
```
